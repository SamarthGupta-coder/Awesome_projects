```python
from urllib.request import urlopen
import bs4 # It may happend that sometimes it may not work, so change the kernel
import pandas as pd 
import json
```

```python
csv_file = pd.read_csv('Jokes Challenge/Test/ID.csv')

df = pd.DataFrame(csv_file)
print("DF",df.loc[0:4],'\n')

id = df.loc[0].tolist()
print(id[0])
```

```python
i=0
while(True):
    try:
        print(df.loc[i].tolist()[0])

    except:
        break
   
    i=i+1
```

```python
i=0

new_df4 = pd.DataFrame(columns=['ID', 'Joke'])

while(True):
    try:
        print(df.loc[i].tolist()[0])
        id = df.loc[i].tolist()[0]

        api_url = "http://api.icndb.com/jokes/" + str(id)
        url_result = urlopen(api_url)

        data = url_result.read()

        json_data = json.loads(data)

        insert_data = [{'ID':id, 'Joke':json_data['value']['joke']}] 

        new_df4 = new_df4.append(insert_data,ignore_index=True,sort=False)
        
        print(insert_data)


    except:
        break
   
    i=i+1
```
